```python
#Finding inter quartile range
IQR = Q3-Q1
IQR
```

```python
#Obtaining lower limit and upper limit
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print(low_lim)
print(up_lim)
```

```python
#Finding outliers
outlier=[]
for x in data['Age']:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
outlier           
```
